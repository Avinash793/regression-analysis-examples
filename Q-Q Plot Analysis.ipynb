```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats

import warnings
warnings.filterwarnings("ignore")
```

<h3> Distribution Sample Size </h3>

```python
n = 1000
```

<h3> Draw Samples From Distributions </h3>

```python
bionomial_dis_data = np.random.binomial(n=10,p=0.5,size=n)
poisson_dis_data = np.random.poisson(lam=5,size=n)

uniform_dis_data = np.random.uniform(low=0,high=1,size=n)
exp_dis_data = np.random.default_rng().exponential(scale=1,size=n)
normal_dis_data = np.random.normal(5,5,size=n)

stdnormal_dis_data = np.random.standard_normal(size=n)
tdis_data = np.random.standard_t(df=10,size=n)
chisquared_dis_data = np.random.chisquare(df=10,size=n)
```

<h3> Plot Distributions </h3>

```python
fig, axes = plt.subplots(4, 2, figsize=(11, 7))

axes[0,0].hist(bionomial_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[0,0].set_title("Bionomial Dist")
axes[0,1].hist(poisson_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[0,1].set_title("Poisson Dist")

axes[1,0].hist(uniform_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[1,0].set_title("Uniform Dist")
axes[1,1].hist(exp_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[1,1].set_title("Exponential Dist")
axes[2,0].hist(normal_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[2,0].set_title("Normal Dist")
axes[2,1].hist(stdnormal_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[2,1].set_title("Standard Normal Dist")

axes[3,0].hist(tdis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[3,0].set_title("Student t Dist")
axes[3,1].hist(chisquared_dis_data, bins=200, density=True, color='blue', alpha=0.7)
axes[3,1].set_title("Chi2 Dist")

plt.tight_layout()
plt.show() 
```

<h3> stdnormal_dis_data Q-Q Plots </h3>

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 7))

sm.qqplot(stdnormal_dis_data, line='45', color='skyblue', ax=axes[0])
axes[0].set_title("Normal Q-Q Plot")
axes[0].set_xlabel("Theoretical Quantiles")
axes[0].set_ylabel("Sample Quantiles")

sm.qqplot(stdnormal_dis_data, stats.t, distargs=(10,), line="45", color='skyblue', ax=axes[1])
axes[1].set_title("t-Dist Q-Q Plot")
axes[1].set_xlabel("Theoretical Quantiles")
axes[1].set_ylabel("Sample Quantiles")

sm.qqplot(stdnormal_dis_data, stats.chi2, distargs=(10,), line='45', color='skyblue', ax=axes[2])
axes[2].set_title("Chi2 Q-Q Plot")
axes[2].set_xlabel("Theoretical Quantiles")
axes[2].set_ylabel("Sample Quantiles")

plt.tight_layout()
plt.show() 
```

We can conclude that stdnormal_dis data arises from Standard Normal distribution.

<h3> tdis_data Q-Q Plots </h3>

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 7))

sm.qqplot(tdis_data, line='45', color='skyblue', ax=axes[0])
axes[0].set_title("Normal Q-Q Plot")
axes[0].set_xlabel("Theoretical Quantiles")
axes[0].set_ylabel("Sample Quantiles")

sm.qqplot(tdis_data, stats.t, distargs=(10,), line='45', color='skyblue', ax=axes[1])
axes[1].set_title("t-Dist Q-Q Plot")
axes[1].set_xlabel("Theoretical Quantiles")
axes[1].set_ylabel("Sample Quantiles")


sm.qqplot(tdis_data, stats.chi2, distargs=(10,), line='45', color='skyblue', ax=axes[2])
axes[2].set_title("Chi2 Q-Q Plot")
axes[2].set_xlabel("Theoretical Quantiles")
axes[2].set_ylabel("Sample Quantiles")

plt.tight_layout()
plt.show() 
```

We can conclude that tdis_data arises from Standard t-Distribution with df=10.

<h3> chisquared_dis_data Q-Q Plots </h3>

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 7))

sm.qqplot(chisquared_dis_data, line='45', color='skyblue', ax=axes[0])
axes[0].set_title("Normal Q-Q Plot")
axes[0].set_xlabel("Theoretical Quantiles")
axes[0].set_ylabel("Sample Quantiles")

sm.qqplot(chisquared_dis_data, stats.t, distargs=(10,), line='45', color='skyblue', ax=axes[1])
axes[1].set_title("t-Dist Q-Q Plot")
axes[1].set_xlabel("Theoretical Quantiles")
axes[1].set_ylabel("Sample Quantiles")

sm.qqplot(chisquared_dis_data, stats.chi2, distargs=(10,), line='45', color='skyblue', ax=axes[2])
axes[2].set_title("Chi2 Q-Q Plot")
axes[2].set_xlabel("Theoretical Quantiles")
axes[2].set_ylabel("Sample Quantiles")

plt.tight_layout()
plt.show() 
```

We can conclude that chisquared_dis_data arises from Chi2 distribution with df=10.

<h3> exp_dis_data Q-Q Plots </h3>

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 7))

sm.qqplot(exp_dis_data, line='45', color='skyblue', ax=axes[0])
axes[0].set_title("Normal Q-Q Plot")
axes[0].set_xlabel("Theoretical Quantiles")
axes[0].set_ylabel("Sample Quantiles")


sm.qqplot(exp_dis_data, stats.t, distargs=(10,), line='45', color='skyblue', ax=axes[1])
axes[1].set_title("t-Dist Q-Q Plot")
axes[1].set_xlabel("Theoretical Quantiles")
axes[1].set_ylabel("Sample Quantiles")

sm.qqplot(exp_dis_data, stats.expon, line="45", color='skyblue', ax=axes[2])
axes[2].set_title("Exponential Q-Q Plot")
axes[2].set_xlabel("Theoretical Quantiles")
axes[2].set_ylabel("Sample Quantiles")

plt.tight_layout()
plt.show() 
```

We can conclude that exp_dis_data arises from Exponential distribution.
